## 1.2 Selecting data using `.loc`, `.iloc`, and mapping with a boolean array

`.loc` uses label-based indexing, but you can also pass a boolean array. Labels refer to the values in the index column of a pandas series or pandas DataFrame, or the column names in a pandas DataFrame.

```python
# as pandas DataFrame is a two dimensional data type - we have two dimensions to provide indexes or slices for
# data selection
df.loc[:, 'person_gender']
```

```python
# We can also specify a value for only one dimension
# Here, we're looking for rows (default axis) where the “index” column has a value of 1, 2, or 3
df.loc[[1,2,3]]
```

```python
df.loc[:5, ['person_gender', 'person_income']]
```

```python
# Note that the value for the first dimension (loc) is the actual value in the index column,
# not the index number of the elements in that index.
# Below is an example where the index has been changed to values starting from 1.
df.index = range(1, df.shape[0] + 1)
```

```python
df[:5]
```

```python
# now we are checking how loc works with the same parameters as iloc
df.loc[:5, ['person_gender', 'person_income']]
```

**WARNING!**

Indexing using labels is performed with the stop value included, which is different from how classic slices work in Python!

Indexing with `.iloc` is similar to `.loc`, but here the letter `i` represents an index, so instead of column and row labels, we specify indices.

```python
# row indexes: 0-4, column indexes: 2 and 3
df.iloc[:5, 2:4]
```

```python
# Just like with the NumPy library, we can also manipulate the data in these areas, such as swapping the order of the columns
# although this isn’t usually necessary, since we can retrieve data from the frame in any column order
# Here’s the column order: third, second, and fourth columns (remember that indexing starts at 0?)
df.iloc[:, [2,1, 3]]
```

```python
# However, if we wanted to reorder them in the original grid, we could, for example, use the vector
# of the current column names from the .columns property
df.columns
```

```python
# Let's say we want to swap columns 2 and 3
# For simplicity, we'll store the values of these labels in a Python list
old_columns = list(df.columns)
new_column_order = old_columns[:1] + old_columns[2:0:-1] + old_columns[3:]
new_column_order
```

```python
# Overwrite the variable with the original data frame, a new data frame with a different column order
df = df.loc[:, new_column_order]
```

```python
df.head()
```

**Selecting Data Using Boolean Value Tables**

This operation is typically used in data filtering, where a condition is applied to select data from a frame or series. Here, we can use the operators `|` (logical OR), `&` (logical AND), and `~` for negation.

```python
# returned type - a numpy array of boolean types (a mapping)
print(df['person_gender'] == 'female')

# now we apply the mapping to the data
df[df['person_gender'] == 'female']
```

```python
# mapping with two conditions (& - this is a logical AND operator)
(df['loan_amnt'] > 10000) & (df['person_income'] < 50000)
```

```python
# and now the data by applying the mapping
df[(df['loan_amnt'] > 10000) & (df['person_income'] < 50000)]
```

```python
# and now we're completing this data by using the same condition and its negation
# an additional parenthesis has been added
df[~((df['loan_amnt'] > 10000) & (df['person_income'] < 50000))]
```

```python
# If we want to filter a column based on multiple values, it will be more useful and concise to
# use the isin() method
df['person_education'].isin(['Associate', 'High School'])
```

```python
# Filtering is also possible using a callable object,
# such as a built-in function or an anonymous function (lambda)
# Here are some examples

# Here, we first access the value `str` from this series, then apply the `startswith(‘D’)` method to each
# value in the series, which returns a Boolean array
map_1 = df['loan_intent'].str.startswith('D')

# Here's the same effect, but using `apply` (which maps something to each element of an iterable object) — here we map a function
# to each value in a given data series, and additionally pass the argument ‘D’ to the mapped function
map_2 = df['loan_intent'].apply(str.startswith, args='D')

# and yet another example - this time using anonymous function (lambda)
map_3 = df['loan_intent'].map(lambda x: x.startswith('D'))

map_1, map_2, map_3
```

```python
# Let's consider a scenario where we're interested in records where, after subtracting the loan amount
# from income, the remaining amount is greater than 50,000

# We can calculate this fairly easily
print(df['person_income'] - df['loan_amnt'])

# now we can use it as a condition (mapping)
netto_income_mapping = (df['person_income'] - df['loan_amnt']) > 50000
print(netto_income_mapping)

# and apply to the real data
df[netto_income_mapping]

# Can this also be done using a lambda function?

# A lambda function receives a parameter for each value returned by the iterator for the iterable
# on which it is called. In the previous case, it was a single data series, so the anonymous function received
# a single value from the series. Now let’s try another example where the function receives multiple values.

# Here, the lambda function’s argument is the entire record from the data frame; we can index it in any way
# that was presented earlier
df.apply(lambda x: (x['person_income'] - x['loan_amnt']) > 50000, axis=1)
```

## 1.3 Using `DataFrame.query`

In addition to the examples shown above, there is also the `query` method, which can be easier to write when filtering large sets of elements.

Examples in the documentation: https://pandas.pydata.org/docs/user_guide/indexing.html#the-query-method

Here are a few examples.

```python
df.query('person_income > 100000')
```

```python
# Here, too, we can use the operators &, |, and ~
# As a reminder
# & - logical AND
# | - logical OR
# ~ - negation
display(df.query('person_education == "Doctorate" & person_age > 40'))
# code below 'person_education == ["Doctorate", "Master"] works like isin(["Doctorate", "Master"])
df.query('person_education == ["Doctorate", "Master"] & person_age > 40')
```
