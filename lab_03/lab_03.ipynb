## 1. Selected arithmetic operations on a NumPy array

```python
import numpy as np
```

```python
# Functions taken from https://tobiasraabe.github.io/post/numpy-views-vs-copies/
# These functions make it easier to determine whether two NumPy arrays have a common base
# which means they share memory (or not)

def get_data_base(x):
    base = x
    while isinstance(base.base, np.ndarray):
        base = base.base
    return base

def arrays_share_data(x, y):
    return get_data_base(x) is get_data_base(y)
```

```python
arr = np.arange(1,101).reshape((10,10))
arr
```

```python
# sum of all values in the array
print(arr.sum())
# lub
print(np.sum(arr))
```

```python
# sum in each column (row indexing)
print(arr.sum(axis=0))
# and each rows (column indexing)
print(arr.sum(axis=1))
```

```python
# cumulative sum
print(arr.cumsum())
```

```python
# most of the functions can be run for a given axis
print(arr.cumsum(axis=0))
print(arr.cumsum(axis=1))
```

```python
# For the purposes of a few examples, we will use a copy of the original array
arr_nan = arr.copy()
```

```python
# Filtering an array returns a Boolean array—a map—of values that satisfy and do not satisfy the condition
divide_by_five_map = (arr_nan % 5 == 0)
divide_by_five_map
```

```python
# We'll also check from time to time whether we're working on a copy or a view
# Here we're dealing with what's known as “fancy slicing”—so it's a copy
arrays_share_data(arr_nan, divide_by_five_map)
```

```python
# We can attach this panel to another panel (make sure the dimensions match!)
divide_by_five = arr_nan[divide_by_five_map]
divide_by_five
```

```python
arrays_share_data(divide_by_five, arr_nan)
```

```python
# We convert the values associated with the map indices to values such as “nan” (not a number)
arr_nan[divide_by_five_map] = np.nan
```

```python
arr_nan
```

```python
arr_nan.dtype
```

```python
# It turns out that we can't do this for types like `int`
# Maybe for types like `float`
# After a failed insertion operation, we should check whether the array contains valid data
# If not, the copying operation should be performed again
arr_nan = arr_nan.astype(np.float64)
arr_nan
```

```python
arr_nan[divide_by_five_map] = np.nan
```

```python
# Let's perform the sum operation again
print(arr_nan.sum())
```

```python
print(arr_nan.sum(axis=0))
print(arr_nan.sum(axis=1))
```

```python
# If we want to treat nan values in a specific way, e.g., as 0 (zero), we have
# dedicated functions at our disposal—in this case, the sum function
print(np.nansum(arr_nan))
print(np.nansum(arr_nan, axis=0))
print(np.nansum(arr_nan, axis=1))
```

```python
# Multiplying array elements
# Not working? Why 0? This is due to an overflow of values for types supported by NumPy.
# Python's built-in int type can represent extremely large numbers (limited only by RAM), but  numpy
# is based on types known from the C language. Value overflow for types like int64 or float64 during multiplication can
# occur very quickly. In this case, 0 is returned without any warning—keep this in mind.
print(arr.prod())
print(arr.prod(axis=0))
```

```python
# NumPy has a wide range of built-in functions for arithmetic, trigonometry, linear algebra, and more.
# We’ll cover linear algebra a bit later. Below are a few examples of other common operations.
print(arr.max(), arr.max(axis=0))
```

```python
# We can also use operations of this type to filter an array
arr[(arr == arr.max(axis=0))]
```

```python
# Here, it’s also worth introducing a function like `where`, which, instead of returning values that meet a condition,
# can return the indices where those conditions are met. Below, we’ll see two arrays, since the array
# we’re filtering is two-dimensional. The first dimension is the row, and the second is the column.
# The `where` function takes up to three arguments: condition, x, y. If only the condition is provided, it returns the indices of the elements
# that satisfy that condition.
```

```python
np.where(arr == arr.max(axis=0))
```

```python
# What is the equivalent of the following?
np.asarray(arr == arr.max(axis=0)).nonzero()
```

```python
# but we can also use it this way
# condition, value if the condition is met, value if the condition is not met
np.where(arr == arr.max(axis=0), arr, np.nan)
```

## 2. `np.random` module

The `np.random` module provides many functions that allow you to generate arrays of random values (more precisely, pseudo-random values, due to the way computers work) with specific properties, as well as to sample from existing arrays.

### 2.1 Selected examples of initializing arrays using `np.random`

To use the module, start by initializing the pseudorandom number generator.

```python
rng = np.random.default_rng()
```

```python
rng.random()
```

```python
rng.random((5,5))
```

```python
# The values returned are in the range [0.0, 1.0)
```

```python
# We can sample pseudo-random values from distributions, such as the normal distribution
# Function signature: normal(loc=0.0, scale=1.0, size=None) (mean, standard deviation, size)
print(rng.normal())
print(rng.normal(size=(5,5)))
```

```python
# or we can use the following function, where (mean=0.0, std=1.0)
rng.standard_normal(5)
```

```python
# Generating integers from a range
rng.integers(low=1, high=100, size=10)
```

```python
# choice - retrieves a random sample of data from an array
# signature: choice(a, size=None, replace=True, p=None, axis=0, shuffle=True)
# The p parameter allows you to pass an array containing a probability distribution
```

```python
# Note the output: a random line is returned
rng.choice(arr)
```

```python
# By default, the selection is made with replacement (replace=True)
rng.choice(arr, size=20)
```

```python
rng.choice(arr, size=5, replace=False)
```

```python
# But what if we wanted to draw a single value from the array?
# We can use the functions we learned in the previous lab
rng.choice(arr.flat, size=10, replace=False)
```

```python
# The array is “shuffled” in-place
rng.shuffle(arr)
arr
```

```python
# and by default axis=0, but we can change that
rng.shuffle(arr, axis=1)
arr
```

```python
# Generate a permutation from an array or by specifying an integer (works with most similar NumPy functions)
# Returns a permutation of integers within the specified range.
# The main difference between `shuffle` and `permutation` is that the former operates in-place
# while the latter returns a copy
rng.permutation(10)
```

```python
rng.permutation(arr)
```
