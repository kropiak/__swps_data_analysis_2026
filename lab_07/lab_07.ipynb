The code above is written in the format typically used in traditional Python scripts. Jupyter Notebook has a special backend and magic methods that usually have additional functionality in notebooks. The `matplotlib` package can also be used with this backend (more here: [https://ipython.readthedocs.io/en/stable/interactive/plotting.html](https://ipython.readthedocs.io/en/stable/interactive/plotting.html).
This frees us from having to use `plt.show()`, but it also prints a string representation of the last line of code (the object) to the console. You can avoid this by assigning this statement to a variable or adding a `;` at the end of the code.

**Listing 2**

```python
# activation of matplotlib inline backend
%matplotlib inline
plt.ylabel('y axis label')
plt.plot([1, 2, 3, 4]);
```

The vector of the passed values is the Y-axis, and the X-axis was generated automatically; here, for each value in the Y-vector, it simply takes the corresponding index value from this list—that is, for the value 1, it takes the value 0, and so on.

**Listing 3**

```python
# We pass two vectors of values, first for the x-axis, then for the y-axis
# Additionally, we have a string parameter here that specifies the chart style
# For a complete list, see the documentation at
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro-')

# below we set up a limit of values for each axis [xmin, xmax, ymin, ymax]
plt.axis([0, 6, 0, 20])
plt.title('Chart 1')
plt.show()

# We can also set different colors for individual elements by overlaying two charts
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o')
plt.title('Chart 2')

plt.axis([0, 6, 0, 20])
```

**Listing 4**

```python
import numpy as np

# base value vector
t = np.arange(0., 5., 0.2)

# With a single call to the plot() function, we can generate multiple plots on a single canvas
# each time specifying the necessary values: values for the x-axis, values for the y-axis, plot style, ...
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^');
```

**Listing 5**

```python
x = np.linspace(0, 2, 100)

# Plots can also be added to the canvas one by one instead of using a single call to the plot() function
# Here, the label parameter is also used to specify the label for a given plot in the legend
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# axis labels
plt.xlabel('x')
plt.ylabel('f(x)')

# plot title
plt.title("Simple line plot")

# enables plot legend
plt.legend();
```

**Listing 6**

### 2.2. Scatter plot

> **Docs**: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

**Listing 7**

```python
# data in dictionary format, but it could just as easily be a Pandas DataFrame
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# To pass the plot parameters in this way, you must include the required `data` parameter, which contains data accessible via labels
# This means that ‘a’ is equivalent to `data[‘a’]`, and so on. The `c` parameter is short for `color`, passed here as a vector
# color values for each successive value of the plot. The parameter s stands for scale—it specifies the size, in this case of the point, for
# each successive value of the plot vector. In summary, the following values will be used for the first point of the plot
print(f"a={data['a'][0]}, b={data['b'][0]}, c={data['c'][0]}, d={data['d'][0]}")
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('a')
plt.ylabel('b');
```

### 2.3 Subplots

Subplots allow you to place multiple plots on a single canvas, arranged in a grid. You specify the grid dimensions, i.e., the number of rows and the number of columns. This is done using the `subplot` function, which takes three arguments (`nrows, ncols, index`). These correspond to the number of grid rows, the number of columns, and the index of the plot being defined (indices start at 1 and end at `nrows*ncols`).

**Listing 8**

```python
x1 = np.arange(0.0, 2.0, 0.02)
x2 = np.arange(0.0, 2.0, 0.02)

y1 = np.sin(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, '-')
plt.title('Two subplots')
plt.ylabel('sin(x)')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r-')
plt.xlabel('x')
plt.ylabel('cos(x)');
```

**Listing 9**

```python
# another example

x1 = np.arange(0.0, 2.0, 0.02)
x2 = np.arange(0.0, 2.0, 0.02)
y1 = np.sin(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)

# indicates that this is a 3x2 grid (3 rows, 2 columns) and we are defining a chart with index 1
plt.subplot(3, 2, 1)
plt.plot(x1, y1, '-')
plt.title('Two subplots')
plt.ylabel('sin(x)')

# We can also omit the commas; here, the graph has an index of 4, so we omitted 2 and 3
plt.subplot(324)
plt.plot(x2, y2, 'r-')
plt.xlabel('x')
plt.ylabel('cos(x)')

plt.subplot(3, 2, 5)
plt.plot(x2, y2, 'r-')
plt.xlabel('x')
plt.ylabel('cos(x)');
```

### 2.6 Pie chart

> **Docs**: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

**Listing 12**

```python
players = ['Messi', 'Suarez', 'Dembele', 'Coutinho']
goals = [48, 25, 13, 11]

# firtst version of the plot
wedges, texts, autotexts = plt.pie(goals, labels=players,
                                   autopct=lambda pct: "{:.1f}%".format(pct), textprops=dict(color="black"))
plt.setp(autotexts, size=12, weight="bold")
plt.title("First version of the plot")
plt.legend(title='Players')
plt.show()

# second version of the plot
def prepare_label(pct, br):
    absolute = int(np.ceil(pct / 100. * np.sum(br)))
    return "{:.1f}% \n({}/{})".format(pct, absolute, sum(goals))


wedges, texts, autotexts = plt.pie(goals, labels=players,
                                   autopct=lambda pct: prepare_label(pct, goals), textprops=dict(color="black"))
plt.setp(autotexts, size=12, weight="bold")
plt.title("Second version of the plot")

# you can learn more about available legend parameters from docs:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(title='Players', loc='best');
```

### 3. Plotting directly from pandas DataFrame

The matplotlib library is a low-level library compared to the libraries that will be introduced in future classes (seaborn, bokeh, plotly) and requires a bit more work when preparing and selecting data for a plot. The example will be based on the data from the `orders.csv` file.

```python
import pandas as pd

df_ord = pd.read_csv(r'https://raw.githubusercontent.com/kropiak/__swps_data_analysis_2026/refs/heads/main/lab_06/data/orders.csv', sep=';')
df_ord.info()
```

```python
# basic line plot for a single column
df_ord['OrderValue'].plot()
```

```python
# or passing a pandas Series to a plot function
plt.plot(df_ord['OrderValue'])
```

```python
# Let’s now consider an example in which we want to display two data series (two lines on a single chart) for each country individually
plt.plot(df_ord[df_ord['Country'] == 'Polska']['OrderValue'])
plt.plot(df_ord[df_ord['Country'] == 'Niemcy']['OrderValue'])
plt.legend(['Polska','Niemcy'])
```

```python
# Now we'd like to do the same thing, but for each year (and country) separately, calculating the total revenue
# The easiest way is to prepare the data in the form of a grouped table with aggregation

# Step 1 – We’ll convert the Order Date column to a datetime type, which will make it much easier to work with the date components (and, for example, group by year, month, etc.)
df_ord['OrderDate'] = pd.to_datetime(df_ord['OrderDate'])
df_ord.info()
```

```python
# Step 2 - We group the data and calculate the total for each group (i.e., country and year)
df_ord_sum_by_year = df_ord.groupby([df_ord['OrderDate'].dt.year, 'Country']).agg({'OrderValue': ['sum']})
df_ord_sum_by_year
```

```python
# Step 3 - Plot
# Let's just try plotting the data for the ‘OrderValue’ and ‘sum’ columns (since we have a multi-index here)
fig = plt.figure(figsize=(10,8))
df_ord_sum_by_year[('OrderValue', 'sum')].plot()

# It doesn't quite look the way we want it to (we want two separate series for the country, with the year on the x-axis)
```

```python
# So let's try doing this separately
# What indexes do we have in the frame?
df_ord_sum_by_year.index
```

```python
# single data point looks like follows
df_ord_sum_by_year.loc[2003,'Polska']
```

```python
# Extracting data from such a frame is quite tricky
# Resetting the index might help
df_ord_sum_by_year_2 = df_ord_sum_by_year.reset_index()
df_ord_sum_by_year_2
```

```python
# now it should be easier
x = df_ord_sum_by_year_2['OrderDate'].unique()
plt.xticks(df_ord_sum_by_year_2['OrderDate'].unique())
plt.xlabel('Year')
plt.plot(x, df_ord_sum_by_year_2[df_ord_sum_by_year_2['Country'] == 'Polska'][('OrderValue','sum')])
plt.plot(x, df_ord_sum_by_year_2[df_ord_sum_by_year_2['Country'] == 'Niemcy'][('OrderValue','sum')])
```

```python
# To organize the data frame so that it's even easier to access the data series, we'll use the unstack() function
df_ord_sum_by_year.unstack()
```

```python

df_ord_sum_by_year_unstacked = df_ord_sum_by_year.unstack()
df_ord_sum_by_year_unstacked.plot()
plt.legend([col[-1] for col in df_ord_sum_by_year_unstacked.columns])
```

```python
# adjusting the x axis
plt.xticks(df_ord_sum_by_year_unstacked.index)
plt.plot(df_ord_sum_by_year_unstacked)
plt.legend(df_ord_sum_by_year_unstacked.columns)
```

```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# data preparation
x = np.linspace(-10, 10, 1000)
init_amplitude = 1.0
init_frequency = 1.0
init_phase = 0.0

def sine_wave(x, amp, freq, phase):
    return amp * np.sin(freq * x + phase)

# main plot window
fig, ax = plt.subplots(figsize=(8, 6))
plt.subplots_adjust(bottom=0.35) # space for sliders at the bottom
line, = ax.plot(x, sine_wave(x, init_amplitude, init_frequency, init_phase), lw=2)
ax.set_ylim(-5, 5)
ax.set_xlim(-10, 10)
ax.grid(True)
ax.set_title('Interactive plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

# configuring axis for sliders (x, y, width, height)
ax_amp = plt.axes([0.15, 0.2, 0.65, 0.03])
ax_freq = plt.axes([0.15, 0.1, 0.65, 0.03])
ax_phase = plt.axes([0.15, 0.0, 0.65, 0.03])

# sliders
slider_amp = Slider(ax_amp, 'Aplitude', 0.1, 5.0, valinit=init_amplitude)
slider_freq = Slider(ax_freq, 'Frequency', 0.1, 5.0, valinit=init_frequency)
slider_phase = Slider(ax_phase, 'Phase', -np.pi, np.pi, valinit=init_phase)

# function that updated plot after slider changes his state
def update(val):
    amp = slider_amp.val
    freq = slider_freq.val
    phase = slider_phase.val
    line.set_ydata(sine_wave(x, amp, freq, phase))
    fig.canvas.draw_idle()

# connecting events with sliders
slider_amp.on_changed(update)
slider_freq.on_changed(update)
slider_phase.on_changed(update)

plt.show()
```
