## 1. `pandas` data structures

The basic data structures in the pandas library are the dataframe, an object of type `DataFrame`, and the series, `Series`, which constitutes a single column of a dataframe. Underlying these structures are NumPy arrays, which we have already encountered in previous classes. Each column in a dataframe is defined by the data type stored in it, including cases where missing values can also be stored (the `Nan` type—not a number).

In pandas, a DataSeries is a structure that, in addition to data in the form of a vector, also contains an index, which can be of any type, provided it is a hashable type in Python. Below are a few examples of how to initialize a DataSeries.

````python
# How to view the docstring for the pandas.Series class without having to search for it online
print(pd.Series.__doc__)
````

**Example 1**

````python
# initialization of an empty pandas series
series_1 = pd.Series()
series_1
````

````python
# main descriptive statistics, we can run this function for a DataFrame
series_1.describe()
````

**Example 2**

````python
data = np.arange(1,6)
series_2 = pd.Series(data)
series_2
````

````python
# now we can see more as we have a numerical series
series_2.describe()
````

**Example 3**

````python
# pandas series from a Python dictionary
data = {key: val for key,val in zip('abcd',[1,2,3,4])}
print(f"Dict: {data}")
series_3 = pd.Series(data)
series_3
````

````python
series_3.index
````

**Example 4**

````python
# we can convert pandas series to a python list
# methods tolist() and to_list()
series_3.tolist()
````

````python
series_3.index.tolist()
````

**Example 5**

````python
# Looking at the docstring for Series, we can see that arithmetic operations on data series are also possible
# provided that certain rules are followed (the length of the data series vector)
series_3 + series_3
````

````python
series_3 ** 2
````

````python
series_3
````

**Example 6**

````python
# We can also take advantage of certain features of the NumPy library related to returning copies and views of arrays
# Below is an example showing how to create a data series that is a view of a NumPy array
arr = np.arange(1, 6)
print(arr)
series_4 = pd.Series(arr, copy=False)
series_4
````

````python
# The `values` attribute returns the series data as a NumPy array
# We compare the addresses of both arrays—if they match, they point to the same memory location
series_4.values.base == arr.base
````

````python
# This allows us to share this memory, along with all its benefits
# and drawbacks (less memory allocated, but no data isolation
# in the event of modifications made through any of the views)

# Modifying the variable naturally updates its value across all views
arr[0] = 99
print(arr)
print(series_4)
````

**Example 7**

````python
# You can name a data series, although this is not often done in practice
series_4.name = "age"
series_4
````

**Example 8**

````python
# We generate a time series using the `date_range(start, number_of_periods, interval)` function
timeseries = pd.date_range('2025-03-20 12:00:00', periods=6, freq='h')
timeseries
````

````python
# The pandas series also has a special index type for time series data
# Here we'll generate such data, but it can also be loaded from files
data = [f'event {n}' for n in range(1, 7)]
timeseries = pd.Series(data, index=timeseries)
timeseries
````

````python
timeseries.index
````

**Example 11**

````python
# There is an overloaded DataFrame constructor that, among other things, allows you to convert a nested list
# (in this case, a two-level list) into a DataFrame without any additional transformations
nested_list = [[1,2,3,4] for _ in range(4)]
df_2 = pd.DataFrame(nested_list, columns=list('ABCD'))
df_2
````

````python
# `describe` works for all numeric columns
df_2.describe()
````

**Example 12**

````python
df_2['D'] = df_2['D'].astype(str)
df_2.describe()
````
